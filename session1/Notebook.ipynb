##### Exercise 1

Simulate a Stochastic Volatility model in python

```python
stochastic_volatility = StochasticVolatility()
stochastic_volatility.run_simulation()
```

Plotting the results of the simulation

```python
sv_observations = stochastic_volatility.observation_history
sv_latent = stochastic_volatility.latent_variables_history
sv_latent_exp = np.exp(sv_latent)

plot_simulation(
    sv_observations, 
    sv_latent_exp,
    title = "Simulation of a Stochastic Volatility Model",
    latent_label = "Exp of Latent Variables (X)",
    fig_size = None
)
```
